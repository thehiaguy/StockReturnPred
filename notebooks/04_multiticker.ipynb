# Multi-Ticker Testing

Runs the same linear regression pipeline on four tickers — AAPL, MSFT, GOOGL, SPY — to test whether the signal generalises beyond a single stock.

The pipeline for each ticker:
1. Download 5 years of daily price data via yfinance
2. Compute returns, lagged returns (Lag1–3), rolling volatility (5/30-day), and momentum (5/30/252-day)
3. 80/20 train/test split
4. Linear regression via Normal Equation
5. Compute Test R², Train R², cumulative strategy return, and win rate

If the model only worked on AAPL it would suggest the features were overfit to one stock's quirks. Consistent performance across all four tickers — including SPY (the S&P 500 ETF itself) — confirms the feature set captures a real and generalisable signal.

## Conclusions — Multi-Ticker Testing

| Ticker | Train R² | Test R² | Cumulative Return | Win Rate |
|--------|----------|---------|-------------------|----------|
| AAPL   | 0.465    | 0.492   | 3.44x             | 71.6%    |
| MSFT   | 0.499    | 0.456   | 4.33x             | 75.1%    |
| GOOGL  | 0.472    | 0.531   | 7.74x             | 77.1%    |
| SPY    | 0.473    | 0.537   | 1.44x             | 75.6%    |

### What I Learned

- **The signal generalises across all four tickers.** Test R² between 0.45 and 0.54 for every ticker — including SPY — means the momentum and volatility features are capturing something real about how prices move, not just an AAPL-specific quirk.

- **The train/test gap is negligible for all tickers** (~0.01–0.04). The model is not overfitting to any individual stock's training period.

- **Cumulative return varies by volatility, not by model quality.** GOOGL returned 7.74x vs SPY's 1.44x despite similar Test R² scores. GOOGL moves ~1.5% per day on average; SPY moves ~0.3%. Same directional accuracy applied to larger daily moves produces much larger compounded gains.

- **SPY generalising confirms the signal is market-wide.** The S&P 500 index itself responds to the same momentum and volatility features as individual stocks, suggesting these patterns are structural properties of equity markets rather than stock-specific behaviour.
